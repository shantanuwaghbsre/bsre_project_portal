```python
from docx import Document
from docx.text.run import Run
import datetime
from docx.enum.text import WD_ALIGN_PARAGRAPH, WD_COLOR_INDEX
from docx.shared import Pt, RGBColor
d = Document(r"D:\bsre_project_portal\Python\.venv\templates\Self Declaration-Certificate1- ugvcl.docx")
```

```python
context = {
    "consumer_number" : "consumer_number",
    "discom" : "discom",
    "date" : "date",
    "guvnl_registration_number" : "guvnl_registration_number",
    "customer_name" : "customer_name",
    "consumer_number" : "consumer_number",
    "circle" : "circle",
    "division" : "division",
    "sub" : "sub",
    "address" : "address",
    "current_sanctioned_load" : "current_sanctioned_load",
    "current_phase" : "current_phase",
    "total_kilowatts" : "total_kilowatts",
    "installation_phase" : "installation_phase",
    "panel_indian" : "panel_indian",
}

for p in d.paragraphs:
    placeholder_count = p.text.count("{{")
    if placeholder_count:
        whole_para_text = p.text
        p.clear()
        for i in range(placeholder_count):
            split_para = whole_para_text.split("{{ ", 1)
            if len(split_para[0]):
                before_para = p.add_run(split_para[0])
                before_para.bold = False
            # print(split_para)
            try:
                styled_run = p.add_run(context[split_para[1].split(" }}", 1)[0]])
                styled_run.bold = True
                styled_run.underline = True
                styled_run.font.size = Pt(12)
            except:
                pass
            try:
                whole_para_text = split_para[1].split(" }}", 1)[1]
            except:
                # print("split_para - ", split_para)
                break
        after_para = p.add_run(whole_para_text)
        after_para.bold = False
```
